```python
def get_str(x):
    if '/' in x:
        return x.replace('/', ' ')
    else:
        return x

features = ['CKG_KND_ACTO_NM']
df[features] = df[features].applymap(get_str)
```

```python
df[['CKG_MTH_ACTO_NM', 'CKG_STA_ACTO_NM', 'CKG_MTRL_ACTO_NM', 'CKG_KND_ACTO_NM']].head(10)
```

```python
def create_soup(x) :
  return ''.join(x['CKG_MTH_ACTO_NM']) + ' ' + ''.join(x['CKG_STA_ACTO_NM']) + ' ' + ''.join(x['CKG_MTRL_ACTO_NM']) + ' ' + ''.join(x['CKG_KND_ACTO_NM'])
df['soup'] = df.apply(create_soup, axis=1)
```

```python
df['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer()
count_matrix = count.fit_transform(df['soup'])
```

```python
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim
```

```python
df = df.reset_index()
indices = pd.Series(df.index, index=df['CKG_NM'])
```

```python
indices
```

```python
def get_recommendations(CKG_NM, cosine_sim=cosine_sim) :
  choice = []

  idx = indices[CKG_NM]

  sim_scores = list(enumerate(cosine_sim[idx]))

  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  sim_scores = sim_scores[1:11]

  food_indices = [i[0] for i in sim_scores]

  return df['CKG_NM'].iloc[food_indices]
```

```python
get_recommendations('현미호두죽')
```
